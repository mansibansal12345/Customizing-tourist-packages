```python
# Simulating the customer data

import random

# Define the number of customers
num_customers = 30

# Define the activities
activities = ['Spa Treatment', 'Yoga Class', 'Meditation Session', 'Beach Relaxation', 'Nature Walk',
              'Skydiving', 'Mountain Trekking', 'Bungee Jumping', 'Scuba Diving', 'Ziplining']

# Define transportation, meal, insurance, and stay options
transport_options = ['Bus', 'Train', 'Flight']
meal_options = ['Vegetarian', 'Non-Vegetarian', 'Vegan']
insurance_options = ['Medical', 'Accidental']
stay_options = ['Hotel', 'Resort', 'Guesthouse']

# Function to normalize a dictionary so its values sum to 1
def normalize_dict(input_dict):
    total = sum(input_dict.values())
    return {k: round(v / total, 4) for k, v in input_dict.items()}

# Function to generate random data for each customer
def generate_customer_data(customer_id):
    # Random duration of trip (in days, range from 3 to 6 days)
    trip_duration = random.randint(3, 6)

    # Random budget for each customer (between 3000 and 6000)
    budget = random.randint(3000, 6000)

    # Happiness index for each activity (random between 0 and 1)
    happiness_activities = normalize_dict({activity: random.uniform(0, 1) for activity in activities})

    # Happiness index for each transportation option (random between 0 and 1)
    happiness_transport = normalize_dict({transport: random.uniform(0, 1) for transport in transport_options})

    # Happiness index for each meal option (random between 0 and 1)
    happiness_meal = normalize_dict({meal: random.uniform(0, 1) for meal in meal_options})

    # Happiness index for each insurance option (random between 0 and 1)
    happiness_insurance = normalize_dict({insurance: random.uniform(0, 1) for insurance in insurance_options})

    # Happiness index for each stay option (random between 0 and 1)
    happiness_stay = normalize_dict({stay: random.uniform(0, 1) for stay in stay_options})

    # Return the generated data
    customer_data = {
        'customer_id': customer_id,
        'trip_duration': trip_duration,
        'budget': budget,
        'happiness_activities': happiness_activities,
        'happiness_transport': happiness_transport,
        'happiness_meal': happiness_meal,
        'happiness_insurance': happiness_insurance,
        'happiness_stay': happiness_stay
    }

    return customer_data

# Generate data for all customers
customers_data = [generate_customer_data(i) for i in range(1, num_customers + 1)]

print(customers_data[0])
```

```python

# Initialize Pyomo Model
model = ConcreteModel()

# Sets (adjusting for only one customer)
model.Customers = Set(initialize=[1])  # Only customer 1
model.Activities = Set(initialize= activities)  # List of activities
model.Stays = Set(initialize=stay_options)
model.Meals = Set(initialize=meal_options)
model.Insurance = Set(initialize=insurance_options)
model.Transport = Set(initialize=transport_options)
model.Days = Set(initialize=range(1, 7))  # Max 6 days based on trip duration

# Parameters (just for customer 1)
activity_cost = cost_data['activity_cost']
stay_cost = cost_data['stay_cost']
meal_cost = cost_data['meal_cost']
insurance_cost = cost_data['insurance_cost']
transport_cost = cost_data['transport_cost']

# Bandwidths (for simplicity, assuming they are available and do not change)
activity_bandwidth = bandwidth_data['activity_bandwidth']
stay_bandwidth = bandwidth_data['stay_bandwidth']
transport_bandwidth = bandwidth_data['transport_bandwidth']

# Customer-specific data for customer 1 (assuming the data is available)
def create_param_dict(param):
    return {1: customers_data[0][param]}  # Only for customer 1

# Parameters using corrected initialization
model.TripDuration = Param(model.Customers, initialize=create_param_dict('trip_duration'))
model.Budget = Param(model.Customers, initialize=create_param_dict('budget'))

# Happiness for Stay Options
model.HappinessStay = Param(
    model.Customers, model.Stays,
    initialize=lambda model, j, s: customers_data[j - 1]['happiness_stay'][s]
)

# Happiness for Meal Options
model.HappinessMeal = Param(
    model.Customers, model.Meals,
    initialize=lambda model, j, m: customers_data[j - 1]['happiness_meal'][m]
)

# Happiness for Insurance
model.HappinessInsurance = Param(
    model.Customers, model.Insurance,
    initialize=lambda model, j, i: customers_data[j - 1]['happiness_insurance'][i]
)

# Happiness for Transport Options
model.HappinessTransport = Param(
    model.Customers, model.Transport,
    initialize=lambda model, j, t: customers_data[j - 1]['happiness_transport'][t]
)

model.HappinessActivity = Param(
    model.Customers, model.Activities,
    initialize=lambda model, j, a: customers_data[j - 1]['happiness_activities'][a] )

# Activity durations
model.ActivityDuration = Param(model.Activities, initialize=lambda i: random.randint(1, 3))  # Random 1-3 hours

# Decision Variables
model.ActivityParticipation = Var(model.Customers, model.Activities, model.Days, within=Binary)
model.StaySelection = Var(model.Customers, model.Stays, within=Binary)
model.MealSelection = Var(model.Customers, model.Meals, model.Days, within=Binary)
model.InsuranceSelection = Var(model.Customers, model.Insurance, within=Binary)
model.TransportSelection = Var(model.Customers, model.Transport, within=Binary)

def total_objective_rule(model):

    happiness_sum = (
        sum(model.HappinessActivity[j, i] * model.ActivityParticipation[j, i, t] for j in model.Customers for i in model.Activities for t in model.Days) +
        sum(model.HappinessStay[j, s] * model.StaySelection[j, s] for j in model.Customers for s in model.Stays) +
        sum(model.HappinessMeal[j, m] * model.MealSelection[j, m, t] for j in model.Customers for m in model.Meals for t in model.Days) +
        sum(model.HappinessInsurance[j, i] * model.InsuranceSelection[j, i] for j in model.Customers for i in model.Insurance) +
        sum(model.HappinessTransport[j, t] * model.TransportSelection[j, t] for j in model.Customers for t in model.Transport)
    )

    return happiness_sum

model.Objective = Objective(rule=total_objective_rule, sense=maximize)

# Constraints

# Bandwidth Constraints
def activity_bandwidth_rule(model, i, t):
    """Ensures the number of customers participating in an activity on a given day doesn't exceed the bandwidth."""
    # Convert activity 'i' to its index if it's a string:
    activity_index = activities.index(i) if isinstance(i, str) else i
    return sum(model.ActivityParticipation[j, i, t] for j in model.Customers) <= activity_bandwidth[activity_index]

model.ActivityBandwidthConstraint = Constraint(model.Activities, model.Days, rule=activity_bandwidth_rule)

def stay_bandwidth_rule(model, s):
    return sum(model.StaySelection[j, s] for j in model.Customers) <= stay_bandwidth[s]

model.StayBandwidthConstraint = Constraint(model.Stays, rule=stay_bandwidth_rule)

def transport_bandwidth_rule(model, t):
    return sum(model.TransportSelection[j, t] for j in model.Customers) <= transport_bandwidth[t]

model.TransportBandwidthConstraint = Constraint(model.Transport, rule=transport_bandwidth_rule)

# Daily Time Limit Constraint
def daily_time_limit_rule(model, j, t):
    return sum(model.ActivityParticipation[j, i, t] * model.ActivityDuration[i] for i in model.Activities) <= 15

model.DailyTimeLimitConstraint = Constraint(model.Customers, model.Days, rule=daily_time_limit_rule)

# Budget Constraint
def budget_constraint_rule(model, j):
    total_cost = (
        sum(activity_cost[activities.index(i)] * model.ActivityParticipation[j, i, t] for i in model.Activities for t in model.Days) +
        sum(stay_cost[s] * model.StaySelection[j, s] for s in model.Stays) +
        sum(meal_cost[m] * model.MealSelection[j, m, t] for m in model.Meals for t in model.Days) +
        sum(insurance_cost[i] * model.InsuranceSelection[j, i] for i in model.Insurance) +
        sum(transport_cost[t] * model.TransportSelection[j, t] for t in model.Transport)
    )
    return total_cost <= model.Budget[j]

model.BudgetConstraint = Constraint(model.Customers, rule=budget_constraint_rule)

# At Least One Activity per Day
def at_least_one_activity_per_day_rule(model, j, t):
    return sum(model.ActivityParticipation[j, i, t] for i in model.Activities) >= 1

model.AtLeastOneActivityPerDayConstraint = Constraint(model.Customers, model.Days, rule=at_least_one_activity_per_day_rule)

def exactly_one_stay_rule(model, j):
    return sum(model.StaySelection[j, s] for s in model.Stays) == 1

model.ExactlyOneStayConstraint = Constraint(model.Customers, rule=exactly_one_stay_rule)

def at_least_one_meal_per_day_rule(model, j, t):
    return sum(model.MealSelection[j, m, t] for m in model.Meals) >= 1

model.AtLeastOneMealPerDayConstraint = Constraint(model.Customers, model.Days, rule=at_least_one_meal_per_day_rule)

# At Least One Insurance Type per Customer
def at_least_one_insurance_rule(model, j):
    return sum(model.InsuranceSelection[j, i] for i in model.Insurance) >= 1

model.AtLeastOneInsuranceConstraint = Constraint(model.Customers, rule=at_least_one_insurance_rule)

# At Least One Transportation Type per Customer
def exactly_one_transportation_rule(model, j):
    return sum(model.TransportSelection[j, i] for i in model.Transport) == 1

model.ExactlyOneTransportConstraint = Constraint(model.Customers, rule=exactly_one_transportation_rule)

def max_activities_per_day_rule(model, j, t):
    return sum(model.ActivityParticipation[j, i, t] for i in model.Activities) <= 3

model.MaxActivitiesPerDayConstraint = Constraint(model.Customers, model.Days, rule=max_activities_per_day_rule)

def no_repeat_activities_across_days_rule(model, j, i):
    # Sum over all days to ensure an activity is selected at most once for each customer
    return sum(model.ActivityParticipation[j, i, t] for t in model.Days) <= 1

model.NoRepeatActivitiesAcrossDaysConstraint = Constraint(model.Customers, model.Activities, rule=no_repeat_activities_across_days_rule)

# Assign higher happiness activities to earlier days: enforce no repeat of activities on different days
def activity_preference_rule(model, j, t, i):
    # Ensure that activities with higher happiness are done earlier
    prev_day = t - 1 if t > 1 else 1
    if t > 1:
        # Prioritize activities with higher happiness on earlier days
        return sum(model.HappinessActivity[j, i] * model.ActivityParticipation[j, i, prev_day] for i in model.Activities) >= model.HappinessActivity[j, i] * model.ActivityParticipation[j, i, t]

    return Constraint.Skip

model.ActivityPreferenceConstraint = Constraint(model.Customers, model.Days, model.Activities, rule=activity_preference_rule)


# Solve the Model for the first customer
solver = SolverFactory('glpk')  # Ensure you have GLPK installed or switch to another solver
result = solver.solve(model, tee=True)

# Check solver status to make sure the model solved correctly
print(f"Solver status: {result.solver.status}")
print(f"Solver termination condition: {result.solver.termination_condition}")

# Displaying the results for Customer 1
if result.solver.status == SolverStatus.ok and result.solver.termination_condition == TerminationCondition.optimal:
    print(f"Results for Customer 1:")

    # Displaying Budget (directly access the parameter)
    budget_value = model.Budget[1]  # No .value needed since it's a parameter
    print(f"  Budget: {budget_value}")


    # Displaying Trip Duration (from the parameter)
    trip_duration_value = model.TripDuration[1]  # Accessing the trip duration for customer 1
    print(f"  Trip Duration: {trip_duration_value} days")

    # Calculating total trip cost
    total_cost = sum(activity_cost[activities.index(i)] * model.ActivityParticipation[1, i, t].value for i in model.Activities for t in model.Days) + \
                 sum(stay_cost[s] * model.StaySelection[1, s].value for s in model.Stays) + \
                 sum(meal_cost[m] * model.MealSelection[1, m, t].value for m in model.Meals for t in model.Days) + \
                 sum(insurance_cost[i] * model.InsuranceSelection[1, i].value for i in model.Insurance) + \
                 sum(transport_cost[t] * model.TransportSelection[1, t].value for t in model.Transport)
    print(f"  Total Trip Cost: {total_cost}")

    # Displaying Activities
    for t in model.Days:
      selected_activities = [i for i in model.Activities if sum(model.ActivityParticipation[1, i, t].value for j in model.Customers) > 0.5]
      print(f"Activities selected for Day {t}: {selected_activities}")
    for t in model.Days:
      selected_meals = [m for m in model.Meals if model.MealSelection[1, m, t].value > 0.5]  # Adjust threshold if necessary
      print(f"Meals selected for Day {t}: {selected_meals}")

    # Displaying Insurance
    selected_insurance = [i for i in model.Insurance if model.InsuranceSelection[1, i].value > 0.5]
    print(f"  Insurance: {', '.join(map(str, selected_insurance)) if selected_insurance else 'None'}")

    # Displaying Transportation
    selected_transport = [t for t in model.Transport if model.TransportSelection[1, t].value > 0.5]
    print(f"  Transportation: {', '.join(map(str, selected_transport)) if selected_transport else 'None'}")

    # Displaying Stay
    selected_stay = [s for s in model.Stays if model.StaySelection[1, s].value > 0.5]
    print(f"  Stay: {', '.join(map(str, selected_stay)) if selected_stay else 'None'}")

```

```python
# Initialize bandwidths
activity_bandwidth = bandwidth_data["activity_bandwidth"].copy()
stay_bandwidth = bandwidth_data["stay_bandwidth"].copy()
transport_bandwidth = bandwidth_data["transport_bandwidth"].copy()

# Iterate over each customer
for customer_id in range(1, 31):
    print(f"\nProcessing Customer {customer_id}")

    model = ConcreteModel()

    # Sets (adjusted for one customer at a time)
    model.Customers = Set(initialize=[customer_id])
    model.Activities = Set(initialize=activities)
    model.Stays = Set(initialize=stay_options)
    model.Meals = Set(initialize=meal_options)
    model.Insurance = Set(initialize=insurance_options)
    model.Transport = Set(initialize=transport_options)
    model.Days = Set(initialize=range(1, 7))

    # Parameters
    def create_param_dict(param):
        return {customer_id: customers_data[customer_id - 1][param]}

    model.TripDuration = Param(model.Customers, initialize=create_param_dict('trip_duration'))
    model.Budget = Param(model.Customers, initialize=create_param_dict('budget'))
    model.HappinessStay = Param(model.Customers, model.Stays, initialize=lambda model, j, s: customers_data[j - 1]['happiness_stay'][s])
    model.HappinessMeal = Param(model.Customers, model.Meals, initialize=lambda model, j, m: customers_data[j - 1]['happiness_meal'][m])
    model.HappinessInsurance = Param(model.Customers, model.Insurance, initialize=lambda model, j, i: customers_data[j - 1]['happiness_insurance'][i])
    model.HappinessTransport = Param(model.Customers, model.Transport, initialize=lambda model, j, t: customers_data[j - 1]['happiness_transport'][t])
    model.HappinessActivity = Param(model.Customers, model.Activities, initialize=lambda model, j, a: customers_data[j - 1]['happiness_activities'][a])
    model.ActivityDuration = Param(model.Activities, initialize=lambda i: random.randint(1, 3))

    # Decision Variables
    model.ActivityParticipation = Var(model.Customers, model.Activities, model.Days, within=Binary)
    model.StaySelection = Var(model.Customers, model.Stays, within=Binary)
    model.MealSelection = Var(model.Customers, model.Meals, model.Days, within=Binary)
    model.InsuranceSelection = Var(model.Customers, model.Insurance, within=Binary)
    model.TransportSelection = Var(model.Customers, model.Transport, within=Binary)

    # Objective
    def total_objective_rule(model):
        return sum(
            model.HappinessActivity[j, i] * model.ActivityParticipation[j, i, t]
            for j in model.Customers for i in model.Activities for t in model.Days
        )
        + 0.1 * sum(  # Reward selecting more activities
            model.ActivityParticipation[j, i, t]
            for j in model.Customers for i in model.Activities for t in model.Days
        ) + sum(
            model.HappinessStay[j, s] * model.StaySelection[j, s]
            for j in model.Customers for s in model.Stays
        ) + sum(
            model.HappinessMeal[j, m] * model.MealSelection[j, m, t]
            for j in model.Customers for m in model.Meals for t in model.Days
        ) + sum(
            model.HappinessInsurance[j, i] * model.InsuranceSelection[j, i]
            for j in model.Customers for i in model.Insurance
        ) + sum(
            model.HappinessTransport[j, t] * model.TransportSelection[j, t]
            for j in model.Customers for t in model.Transport
        )

    model.Objective = Objective(rule=total_objective_rule, sense=maximize)

    # Constraints
    def activity_bandwidth_rule(model, i, t):
    # Get the index of the activity in the activities list
      index = activities.index(i)
    # Use the index to access the activity_bandwidth list
      return sum(model.ActivityParticipation[j, i, t] for j in model.Customers) <= activity_bandwidth[index]

    model.ActivityBandwidthConstraint = Constraint(model.Activities, model.Days, rule=activity_bandwidth_rule)

    def stay_bandwidth_rule(model, s):
        return sum(model.StaySelection[j, s] for j in model.Customers) <= stay_bandwidth[s]

    model.StayBandwidthConstraint = Constraint(model.Stays, rule=stay_bandwidth_rule)

    def transport_bandwidth_rule(model, t):
        return sum(model.TransportSelection[j, t] for j in model.Customers) <= transport_bandwidth[t]

    model.TransportBandwidthConstraint = Constraint(model.Transport, rule=transport_bandwidth_rule)

    # Daily Time Limit Constraint
    def daily_time_limit_rule(model, j, t):
      return sum(model.ActivityParticipation[j, i, t] * model.ActivityDuration[i] for i in model.Activities) <= 15

    model.DailyTimeLimitConstraint = Constraint(model.Customers, model.Days, rule=daily_time_limit_rule)

    # Budget Constraint
    def budget_constraint_rule(model, j):
      total_cost = (
      sum(activity_cost[activities.index(i)] * model.ActivityParticipation[j, i, t] for i in model.Activities for t in model.Days) +
      sum(stay_cost[s] * model.StaySelection[j, s] for s in model.Stays) +
      sum(meal_cost[m] * model.MealSelection[j, m, t] for m in model.Meals for t in model.Days) +
      sum(insurance_cost[i] * model.InsuranceSelection[j, i] for i in model.Insurance) +
      sum(transport_cost[t] * model.TransportSelection[j, t] for t in model.Transport)
    )
      return total_cost <= model.Budget[j]

    model.BudgetConstraint = Constraint(model.Customers, rule=budget_constraint_rule)

    # At Least One Activity per Day
    def at_least_one_activity_per_day_rule(model, j, t):
        return sum(model.ActivityParticipation[j, i, t] for i in model.Activities) >= 1

    model.AtLeastOneActivityPerDayConstraint = Constraint(model.Customers, model.Days, rule=at_least_one_activity_per_day_rule)

    def exactly_one_stay_rule(model, j):
        return sum(model.StaySelection[j, s] for s in model.Stays) == 1

    model.ExactlyOneStayConstraint = Constraint(model.Customers, rule=exactly_one_stay_rule)

    def at_least_one_meal_per_day_rule(model, j, t):
        return sum(model.MealSelection[j, m, t] for m in model.Meals) >= 1

    model.AtLeastOneMealPerDayConstraint = Constraint(model.Customers, model.Days, rule=at_least_one_meal_per_day_rule)

    # At Least One Insurance Type per Customer
    def at_least_one_insurance_rule(model, j):
        return sum(model.InsuranceSelection[j, i] for i in model.Insurance) >= 1

    model.AtLeastOneInsuranceConstraint = Constraint(model.Customers, rule=at_least_one_insurance_rule)

    # At Least One Transportation Type per Customer
    def exactly_one_transportation_rule(model, j):
        return sum(model.TransportSelection[j, i] for i in model.Transport) == 1

    model.ExactlyOneTransportConstraint = Constraint(model.Customers, rule=exactly_one_transportation_rule)

    def max_activities_per_day_rule(model, j, t):
        return sum(model.ActivityParticipation[j, i, t] for i in model.Activities) <= 3

    model.MaxActivitiesPerDayConstraint = Constraint(model.Customers, model.Days, rule=max_activities_per_day_rule)

    def no_repeat_activities_across_days_rule(model, j, i):
        # Sum over all days to ensure an activity is selected at most once for each customer
        return sum(model.ActivityParticipation[j, i, t] for t in model.Days) <= 1

    model.NoRepeatActivitiesAcrossDaysConstraint = Constraint(model.Customers, model.Activities, rule=no_repeat_activities_across_days_rule)

    # Assign higher happiness activities to earlier days: enforce no repeat of activities on different days
    def relaxed_activity_preference_rule(model, j, t, i):
          # Ensure that higher happiness activities are more likely to be selected earlier
          if t > 1:  # Only consider preferences starting from day 2
              prev_day = t - 1
              return (
                  sum(model.HappinessActivity[j, i] * model.ActivityParticipation[j, i, prev_day] for i in model.Activities)
                  >= 0.5 * model.HappinessActivity[j, i] * model.ActivityParticipation[j, i, t]
              )
          return Constraint.Skip

    model.ActivityPreferenceConstraint = Constraint( model.Customers, model.Days, model.Activities, rule=relaxed_activity_preference_rule)
    # Solve the model
    solver = SolverFactory('glpk')
    result = solver.solve(model, tee=False)

    if result.solver.status == SolverStatus.ok and result.solver.termination_condition == TerminationCondition.optimal:
        print(f"Customer {customer_id}: Solution found!")

        # Displaying Budget (directly access the parameter)
        budget_value = model.Budget[customer_id]  # No .value needed since it's a parameter
        print(f"  Budget: {budget_value}")


        # Displaying Trip Duration (from the parameter)
        trip_duration_value = model.TripDuration[customer_id]  # Accessing the trip duration for customer 1
        print(f"  Trip Duration: {trip_duration_value} days")

        # Calculating total trip cost
        total_cost = sum(activity_cost[activities.index(i)] * model.ActivityParticipation[customer_id, i, t].value for i in model.Activities for t in model.Days) + \
                    sum(stay_cost[s] * model.StaySelection[customer_id, s].value for s in model.Stays) + \
                    sum(meal_cost[m] * model.MealSelection[customer_id, m, t].value for m in model.Meals for t in model.Days) + \
                    sum(insurance_cost[i] * model.InsuranceSelection[customer_id, i].value for i in model.Insurance) + \
                    sum(transport_cost[t] * model.TransportSelection[customer_id, t].value for t in model.Transport)
        print(f"  Total Trip Cost: {total_cost}")

        # Displaying Activities
        for t in range(1, int(model.TripDuration[customer_id]) + 1):
          selected_activities = [i for i in model.Activities if sum(model.ActivityParticipation[customer_id, i, t].value for j in model.Customers) > 0.5]
          print(f"Activities selected for Day {t}: {selected_activities}")
        for t in range(1, int(model.TripDuration[customer_id]) + 1):  # Use trip duration
          selected_meals = [m for m in model.Meals if model.MealSelection[customer_id, m, t].value > 0.5]  # Adjust threshold if necessary
          print(f"Meals selected for Day {t}: {selected_meals}")

        # Displaying Insurance
        selected_insurance = [i for i in model.Insurance if model.InsuranceSelection[customer_id, i].value > 0.5]
        print(f"  Insurance: {', '.join(map(str, selected_insurance)) if selected_insurance else 'None'}")

        # Displaying Transportation
        selected_transport = [t for t in model.Transport if model.TransportSelection[customer_id, t].value > 0.5]
        print(f"  Transportation: {', '.join(map(str, selected_transport)) if selected_transport else 'None'}")

        # Displaying Stay
        selected_stay = [s for s in model.Stays if model.StaySelection[customer_id, s].value > 0.5]
        print(f"  Stay: {', '.join(map(str, selected_stay)) if selected_stay else 'None'}")


        # Update bandwidths
        for i in activities:
            index = activities.index(i)  # Get the index of the activity in the activities list
            activity_bandwidth[index] -= sum(model.ActivityParticipation[customer_id, i, t].value for t in model.Days)
        for s in stay_options:
            stay_bandwidth[s] -= model.StaySelection[customer_id, s].value
        for t in transport_options:
            transport_bandwidth[t] -= model.TransportSelection[customer_id, t].value
    else:
        print(f"Customer {customer_id}: Solver failed or no feasible solution.")


# Print remaining bandwidths
print("\nFinal Bandwidths:")
print("Activity Bandwidth:", activity_bandwidth)
print("Stay Bandwidth:", stay_bandwidth)
print("Transport Bandwidth:", transport_bandwidth)
```
